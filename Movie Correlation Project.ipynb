## Numerize non numeric metrics

```python
# look at company -- Correlation with non numeric metrics
df_numerized= df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
df_numerized.corr()
sns.heatmap(df_numerized.corr(), annot=True)
plt.title('Correlation Matric for Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
corr_mat = df_numerized.corr()
corr_pair = corr_mat.unstack()
corr_pair
```

```python
sorted_pairs = corr_pair.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]
high_corr
```
